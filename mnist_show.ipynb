```python
import matplotlib.pyplot as plt
import matplotlib
import numpy as np
from keras.datasets import mnist

%matplotlib inline
matplotlib.rcParams['font.sans-serif'] = ['SimHei']
matplotlib.rcParams['axes.unicode_minus'] = False

#加载数据集
(train_x,train_y),(test_x,test_y)=mnist.load_data()
```

```python
#测试集
X = []
Y = []

for i in range(10):
    x = i
    y = np.sum(test_y == i)
    X.append(x)
    Y.append(y)
    plt.text(x, y, '%s' % y, ha='center', va= 'bottom')

plt.bar(X, Y, facecolor='red', edgecolor='white')
plt.xlabel('数字0-9')
plt.ylabel('测试集集中数量')
plt.xticks(X)
plt.show()
```

```python
#训练集
X = []
Y = []

for i in range(10):
    x = i
    y = np.sum(train_y == i)
    X.append(x)
    Y.append(y)
    plt.text(x, y, '%s' % y, ha='center', va= 'bottom')

plt.bar(X, Y, facecolor='red', edgecolor='white')
plt.xlabel('数字0-9')
plt.ylabel('训练集中数量')
plt.xticks(X)
plt.show()
```

```python
fig, ax = plt.subplots(
    nrows=5,
    ncols=5,
    sharex='all',
    sharey='all', )

ax = ax.flatten()
for i in range(25):
    img = train_x[i].reshape(28, 28)
    ax[i].imshow(img, cmap='Greys', interpolation='nearest')

ax[0].set_xticks([])
ax[0].set_yticks([])
plt.tight_layout()
plt.show()
```
